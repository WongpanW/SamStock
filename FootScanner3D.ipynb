```python
import csv

# Path to the uploaded CSV file
file_path = r'C:\Users\Wongpan\Downloads\GroundTruth-for3D-Subset.csv'

# Initialize an empty list to store the data
data_list = []

# Read data from the CSV file and store it in the list using a loop
with open(file_path, mode='r', newline='', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)  # Extract the header row
    for row in csv_reader:
        row_dict = {}
        for i in range(len(header)):
            row_dict[header[i]] = row[i]
        data_list.append(row_dict)

# Display the first few rows of the data to verify
data_list[:11]
```
